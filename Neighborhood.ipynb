```python
import pandas as pd
import urllib.request
```

```python
wiki='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M' 
```

```python
page=urllib.request.urlopen(wiki)
```

```python
right_table=soup.find('table',{"class": 'wikitable sortable'})
right_table
```

```python
A=[]
B=[]
C=[]
for row in right_table.findAll("tr"):
    cells = row.findAll('td')
    states=row.findAll('th') 
    if len(cells)==3: 
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
       
```

```python

df=pd.DataFrame(A,columns=['Postcode'])
df['Borough']=B
df['Neighborhood']=C
df
```

### Replace Not assigned to NaN(python default )


```python
import numpy as np
df.replace("Not assigned",np.nan,inplace=True)
df.head()
```


###  Drop the NaN datas from the table

```python
df.dropna(subset=['Borough'],axis=0,inplace=True)
df.reset_index(drop=True,inplace=True)

df
```

### Use groupby to merge the rows

```python
df_new=df.groupby(['Postcode','Borough'],as_index=False).agg('sum')
df_new
```

```python
df_new.shape
```

## Code to generate the latitude and longitude:
### Read the csv file and assign it to the new DataFrame

```python
path="http://cocl.us/Geospatial_data"
toronto_data=pd.read_csv(path)
toronto_data.head()
```

```python
toronto_data.rename({'Postal Code':'Postcode'},axis=1,inplace=True)
toronto_data.head()
```

## Merge the two df to get the lat, lng values of borough and neighborhood


```python
toronto_data_new=pd.merge(df_new,toronto_data,how='left',on='Postcode')
toronto_data_new
```

### Generate latitude and longitude of Toronto

```python
address='Toronto, TN'
geolocator=Nominatim(user_agent="TN_explorer")
location=geolocator.geocode(address)
latitude=location.latitude
longitude=location.longitude
print("The geographic coordinates of Toronto are {},{}:".format(latitude,longitude))
```

### Map of Toronto with neighboorhoods superimposed on top

```python
map_toronto=folium.Map(location=[40.4642335,-80.6009058],zoom_start=11)
for lat,lng,borough,neighborhood in zip(toronto_data_new['Latitude'],toronto_data_new['Longitude'],toronto_data_new['Borough'],toronto_data_new['Neighborhood']):
    label='{}, {}'.format(neighborhood,borough)
    label=folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat,lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)
map_toronto
    
```
